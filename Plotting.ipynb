```python
import os
import numpy
import glob
```

```python
distance_file = os.path.join('distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype ='unicode')
headers = distances[0,:]
data = distances[1:,:]
data = data.astype(numpy.float)
```

```python
import matplotlib.pyplot as plt #import as allows to use another name for the library
```

```python
sample = headers[1]
sample2 = headers[2]

plt.plot(data[:,1], label=sample)#plot the second row of data
plt.plot(data[:,2], label=sample2)#you can add a label but it will not show unless you use "plt.legend()"
plt.xlabel('Simulation Frame')#add label on the x-axis called "Simulation Frame"
plt.ylabel('Distance(angstrom)')#add label on the y axis called "Distance(angstrom)"
plt.legend()

plt.savefig('two_samples.png', dpi=300)#save graph as png with resolution 300
```

```python
#Plot all 4 data sets in diffrent graphs and save
for col in range(1, num_columns):
    plt.figure()
    plt.plot(data[:,col], label = headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.legend()
    plt.savefig(F'{headers[col]}.png', dpi = 200)
```
